```python
import numpy as np
```

```python
def read_input(infile):
    with open(infile, 'r') as inf:
        return int(inf.readline().strip())
        
def reduce(n):
    e1 = np.arange(n) + 1

    while len(e1) > 1:
        odd = len(e1) % 2 != 0
        e1 = e1[::2]
        while len(e1) > 1 and odd:  # An odd number of elves
            odd = False
            e2 = e1[1::2]
            if e1[-1] == e2[-1]:
                odd = True
            e1 = e2
    return e1[0]

def reduce2(n):
    e = np.arange(n) + 1

    l = len(e)
    while len(e) > 1:
        l = len(e)
        m3 = l % 3
        m2 = l % 2
        if m3 == 0:
            e2 = e[2::3]
        elif m3 == 1:
            if m2 == 0:
                e2 = np.concatenate((e[(l%2):l//2:3], e[l//2-1:l//2], e[l//2+2::3]))
            else:
                e2 = np.concatenate((e[:l//2-2:3], e[l//2-1:l//2], e[l//2+1::3]))
        elif m3 == 2:
            if m2 == 0:
                e2 = np.concatenate((e[1:l//2:3], e[l//2-1:l//2], e[l//2+2::3]))
            else:
                e2 = np.concatenate((e[1:l//2:3], e[l//2+1:l//2], e[l//2+1::3]))
        e = e2
    return e[0]
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n=========\n')

winner = reduce(5)

print(f'Winner is: {winner}')

assert winner == 3

print('\nPuzzle case\n===========\n')

inp = read_input('input19.txt')

winner = reduce(inp)

print(f'Winner is: {winner}')

assert winner == 1816277

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n=========\n')

winner = reduce2(5)

print(f'Winner is: {winner}')

assert winner == 2

print('\nPuzzle case\n===========\n')

inp = read_input('input19.txt')

winner = reduce2(inp)

print(f'Winner is: {winner}')

assert winner == 1410967
```
