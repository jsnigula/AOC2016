```python
regs = {'a': 0, 'b': 0, 'c': 0, 'd': 0}

def cpy(frm, to, pos):
    if frm in ['a', 'b', 'c', 'd']:
        regs[to] = regs[frm]
    else:
        regs[to] = int(frm)
    return pos + 1

def inc(r, pos):
    regs[r] += 1
    return pos + 1

def dec(r, pos):
    regs[r] -= 1
    return pos + 1

def jnz(r, dist, pos):
    cval = 0
    if r in ['a', 'b', 'c', 'd']:
        cval = regs[r]
    else:
        cval = int(r)

    if cval == 0:
        return pos+1

    return pos + int(dist)

insmap = {'cpy': cpy, 'inc': inc, 'dec': dec, 'jnz': jnz}
```

```python
def read_code(infile):
    instructions = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            if line.strip() != '':
                vals = line.split()
                instructions.append((vals[0], vals[1:]))
    return instructions

def execute(instructions):
    n_ins = len(instructions)
    i = 0
    while True:
        # print(i)
        if i >= n_ins:
            return
        # print('Executing', i, instructions[i], instructions[i][1:]) 
        i = insmap[instructions[i][0]](*instructions[i][1], i)
        
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n=========\n')

regs = {'a': 0, 'b': 0, 'c': 0, 'd': 0}

execute(read_code('input12a.txt'))

print(f'Value in register a is {regs["a"]}')

assert regs['a'] == 42

print('\nPuzzle case\n===========\n')

regs = {'a': 0, 'b': 0, 'c': 0, 'd': 0}

execute(read_code('input12.txt'))

print(f'Value in register a is {regs["a"]}')

assert regs['a'] == 318003

print('\n*******\nPuzzle2\n*******\n')

print('Puzzle case\n===========\n')

regs = {'a': 0, 'b': 0, 'c': 1, 'd': 0}

execute(read_code('input12.txt'))

print(f'Value in register a is {regs["a"]}')

assert regs['a'] == 9227657
```
