```python
def is_wall(x, y):
    if f'{x*x + 3*x + 2*x*y + y + y*y + dnum:b}'.count('1') % 2 == 1:
        return -1
    return 0

def steps_to(tx, ty):
    to_check = {0: [(1, 1)]}
    seen = {(1,1): 0}
    shortest = 0

    while True:
        for px, py in to_check[shortest]:
            for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                nx = px+dx
                ny = py+dy
                if nx < 0 or ny < 0:
                    # Outside
                    continue
                if nx == tx and ny == ty:
                    return shortest+1
                if (nx, ny) not in seen:
                    if is_wall(nx, ny):
                        seen[(nx,ny)] = -1
                    else:
                        seen[(nx, ny)] = shortest + 1
                        if shortest+1 not in to_check:
                            to_check[shortest+1] = []
                        to_check[shortest+1].append((nx, ny))
        shortest += 1

def n_reachable(n):
    to_check = {0: [(1, 1)]}
    seen = {(1,1): 0}
    shortest = 0

    while shortest < n:
        for px, py in to_check[shortest]:
            for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                nx = px+dx
                ny = py+dy
                if nx < 0 or ny < 0:
                    # Outside
                    continue
                if (nx, ny) not in seen:
                    if is_wall(nx, ny):
                        seen[(nx,ny)] = -1
                    else:
                        seen[(nx, ny)] = shortest + 1
                        if shortest+1 not in to_check:
                            to_check[shortest+1] = []
                        to_check[shortest+1].append((nx, ny))
        shortest += 1

    n_pos = 0
    for i in range(0, 51):
        n_pos += len(to_check[i])

    return n_pos

def read_input(infile):
    with open(infile, 'r') as inf:
        return int(inf.readline())
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n=========\n')

dnum = read_input('input13a.txt')

n_steps = steps_to(7, 4)

print(f'Number of steps needed: {n_steps}')

assert n_steps == 11

print('\nPuzzle case\n===========\n')

dnum = read_input('input13.txt')

n_steps = steps_to(31, 39)

print(f'Number of steps needed: {n_steps}')

assert n_steps == 92

print('\n*******\nPuzzle1\n*******\n')

print('\nPuzzle case\n===========\n')

dnum = read_input('input13.txt')

n_pos = n_reachable(50)

print(f'Number of positions reachable in 50 steps: {n_pos}')

assert n_pos == 124

```
