```python
def swap_pos(s, ins):
    p1, _, _, p2 = ins.strip().split()
    p1, p2 = sorted([int(p1), int(p2)])
    return s[0:p1] + s[p2] + s[p1+1:p2] + s[p1] + s[p2+1:]

def r_swap_pos(s, ins):
    return swap_pos(s, ins)

def swap_letter(s, ins):
    l1, _, _, l2 = ins.strip().split()
    p1, p2 = sorted([s.find(l1), s.find(l2)])
    return swap_pos(s, f'{p1} with position {p2}')

def r_swap_letter(s, ins):
    return swap_letter(s, ins)

def reverse_pos(s, ins):
    p1, _, p2 = ins.strip().split()
    p1, p2 = sorted([int(p1), int(p2)])
    return s[0:p1] + ''.join(reversed([c for c in s[p1:p2+1]])) + s[p2+1:]

def r_reverse_pos(s, ins):
    return reverse_pos(s, ins)

def rotate_left(s, ins):
    p1, _ = ins.strip().split()
    p1 = int(p1) % len(s)
    return s[p1:] + s[:p1]

def r_rotate_left(s, ins):
    return rotate_right(s, ins)
    
def rotate_right(s, ins):
    p1, _ = ins.strip().split()
    p1 = -1 * (int(p1) % len(s))
    return s[p1:] + s[:p1]

def r_rotate_right(s, ins):
    return rotate_left(s, ins)
    
def rotate_based(s, ins):
    l = ins.strip().split()[4]
    idx = s.find(l)
    n_rot = idx + 1
    if idx >= 4:
        n_rot += 1
    return rotate_right(s, f'{n_rot} steps')

def r_rotate_based(s, ins):
    n_rot = 0
    while True:
        s_rot = rotate_left(s, f'{n_rot} steps')
        if rotate_based(s_rot, ins) == s:
            return s_rot
        n_rot += 1

def move(s, ins):
    p1, _, _, p2 = ins.strip().split()
    p1, p2 = int(p1), int(p2)
    c = s[p1]
    s = s.replace(c, '')
    return s[0:p2] + c + s[p2:]

def r_move(s, ins):
    p1, _, _, p2 = ins.strip().split()
    p1, p2 = int(p1), int(p2)
    return move(s, f'{p2} to position {p1}')

def process(infile, password):
    s = password
    pmap = {'swap position': swap_pos,
            'swap letter': swap_letter,
            'rotate left': rotate_left,
            'rotate right': rotate_right,
            'rotate based': rotate_based,
            'reverse positions': reverse_pos,
            'move position': move}
    with(open(infile, 'r')) as inf:
        for line in inf.readlines():
            vals = line.strip().split()
            s = pmap[' '.join(vals[0:2])](s, ' '.join(vals[2:]))

    return s

def r_process(infile, password):
    s = password
    pmap = {'swap position': r_swap_pos,
            'swap letter': r_swap_letter,
            'rotate left': r_rotate_left,
            'rotate right': r_rotate_right,
            'rotate based': r_rotate_based,
            'reverse positions': r_reverse_pos,
            'move position': r_move}
    with(open(infile, 'r')) as inf:
        lines = []
        for line in inf.readlines():
            lines.append(line)
        for line in reversed(lines):
            vals = line.strip().split()
            s = pmap[' '.join(vals[0:2])](s, ' '.join(vals[2:]))

    return s
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n=========\n')

s = process('input21a.txt', 'abcde')

print(f'Scrambled password is {s}')

assert s == 'decab'

print('\nPuzzle case\n===========\n')

s = process('input21.txt', 'abcdefgh')

print(f'Scrambled password is {s}')

assert s == 'agcebfdh'

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n=========\n')

s = r_process('input21a.txt', 'decab')

print(f'Original password is {s}')

assert s == 'abcde'

print('\nPuzzle case\n===========\n')

s = r_process('input21.txt', 'fbgdceah')

print(f'Original password is {s}')

assert s == 'afhdbegc'
```
